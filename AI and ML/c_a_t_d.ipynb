```python
df['Rating Count'].fillna(df['Rating Count'].mean(),inplace=True)
```

```python
df.head()
```

```python
df['Rating Count'] = df['Rating Count'].astype(int)
df.info()
```

```python
sns.boxplot(x = 'Rating Count',data = df)
plt.show()
```

```python
# outlier 
q1 = df['Rating Count'].quantile(0.25)
q3 = df['Rating Count'].quantile(0.75)

iqr = q3 - q1
lowerbound = q1 - 1.5 * iqr
upperbound = q3 + 1.5 * iqr

outliers = df[(df['Rating Count'] < lowerbound) | (df['Rating Count'] > upperbound)]
outliers.index

df.drop(outliers.index, inplace=True)

sns.boxplot(x = 'Rating Count',data = df)
plt.show()
```
