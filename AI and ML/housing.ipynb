```python
numeric_cols = ['price','lotsize','bedrooms','bathrms','stories','garagepl']
q1 = df[numeric_cols].quantile(0.25)
q3 = df[numeric_cols].quantile(0.75)
iqr = q3 - q1
lowerbound = q1 - 1.5 * iqr
upperbound = q3 + 1.5 * iqr
outlier = df[((df[numeric_cols] < lowerbound) | (df[numeric_cols] > upperbound)).any(axis=1)] # Takes column by column
df.drop(outlier.index, inplace=True)
```

```python
# Treating of categorical data
cat_cols = ['driveway','recroom','fullbase','gashw','airco','prefarea']
# Lable encoding method
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for cat_col in cat_cols:
    df[cat_col] = le.fit_transform(df[cat_col])
```

```python
df.head()
```

```python
# Feature Scalling -> vaiables not in the range we need to do feature scalling (larger variables)
scaling = ['price','lotsize']
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
for cols in scaling:
    df[cols] = mm.fit_transform(df[[cols]])
```

```python
x = df[['lotsize', 'bedrooms', 'bathrms', 'stories', 'driveway','recroom', 'fullbase', 'gashw', 'airco', 'garagepl', 'prefarea']]
y = df['price']
from sklearn.feature_selection import SelectKBest,f_regression
kbest = SelectKBest(f_regression,k=3) # k -> how many variable you want in the x variable....(best attributes)
kbest.fit_transform(x,y)
kbest.get_support(indices=True) # Column numbers from the selected values
```

```python
# Train test split.....
x = df[['lotsize','airco','bathrms']]
y = df['price']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,shuffle=True,random_state=1)
```

```python
# Simple linear Regression (Model Building)
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(x_train,y_train)
```

```python
slr.coef_ # Coffecient
```

```python
slr.intercept_ # intercept
```

```python
# price = 0.5 * x + 0.17 -> two coffecient and one intercept
```

```python
# Model Evaluation -> happen in the test data | learning done in the train data
ypred = slr.predict(x_test)
```

```python
ypred
```

```python
len(ypred)
```

```python
# Model Extraction
from sklearn.metrics import r2_score
r2_score(y_test,ypred)
```
