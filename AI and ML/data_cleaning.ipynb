```python
# Removal of outliers....
df.describe()
```

```python
Q1 = df['SepalWidthCm'].quantile(0.25)
Q1
```

```python
Q3 = df['SepalWidthCm'].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1; print(IQR)
```

```python
lowerbound = Q1 - 1.5 * IQR
```

```python
upperbound = Q3 + 1.5 * IQR; print(f'{lowerbound} -- {upperbound}')
```

```python
outliers = df[(df['SepalWidthCm'] < lowerbound) | (df['SepalWidthCm'] > upperbound)]
outliers
```

```python
outliers.index
```

```python
df.drop(outliers.index, inplace=True, axis=0)
```
